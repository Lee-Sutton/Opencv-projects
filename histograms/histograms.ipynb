## Color Histograms

```python
plt.imshow(rgb_image)
image = cv2.imread('images/beach.png')
```

```python
chans = cv2.split(image)
colors = ['b','g','r']
plt.figure()

# Run through each of the channels and plot the histogram for each color
color_index = 0
for chan in chans:
    hist = cv2.calcHist([chan], [0], None, [256], [0, 256])
    plt.plot(hist, color=colors[color_index])
    color_index = color_index + 1
```

## 2D color histograms

```python
plt.figure()
hist = cv2.calcHist([chans[1],chans[0]], [0, 1], None, [32, 32], [0, 256, 0, 256,])
plt.imshow(hist)
print hist.shape
```
